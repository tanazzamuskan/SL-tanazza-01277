#Lab 2
Write a python program to calculate mean square and mean absolute error.

```python
def calculate_mse_and_mae(actual, predicted):
    n = len(actual)
    sum_squared_error = 0
    sum_absolute_error = 0
    for i in range(n):
        sum_squared_error += (actual[i] - predicted[i]) ** 2
        sum_absolute_error += abs(actual[i] - predicted[i])
    mse = sum_squared_error / n
    mae = sum_absolute_error / n
    return mse, mae

#defining values and calling functions
actual = [3, -0.5, 2, 7.5]
predicted = [2.5, 0.0, 2, 8]

mse, mae = calculate_mse_and_mae(actual, predicted)
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```

# Lab 3
Write a python program to calculate gradient descent of a machine learning model.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func):
    x_values = np.linspace(-5, 5, 1000)
    y_values = [func(x) for x in x_values]
    print("x values are: ", x_values)
    print("y values are: ", y_values)
    plt.plot(x_values, y_values)
    plt.show()
def plot_derivative(func):
    x_values = np.linspace(-5, 5, 1000)
    delta_x=0.0001
    y_values=(func(x_values+delta_x)-func(x_values))/delta_x
    plt.plot(x_values, y_values)
    plt.show()
#calling functions
plot_function(lambda x: np.sin(x))
plot_derivative(lambda x: np.sin(x))
```

#Lab 7
Prepare a decision tree model for Iris Dataset using Gini Index.

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
df=pd.read_csv("Iris.csv")
print("Top 5 value of dataset:\n",df.head(),"\n----------------------------------------------")
x=df.drop(['Species','Id'],axis=1)
y=df['Species']
print("Top 5 of x:\n",x.head(),"\n--------------------------\nTop 5 of y:\n",y.head())
```

```python
model=DecisionTreeClassifier(criterion='gini')
model
gini_impurity={}
for i in range(x.shape[1]):
  model=model.fit(x.iloc[:,i].values.reshape(-1,1),y) #fitting to x and y as dataframe
  prob=model.predict_proba(x.iloc[:,i].values.reshape(-1,1)) #calculating probabilities all the way down a col
  gini_impurity[i]=1-(prob[:,0]**2+prob[:,1]**2+prob[:,2]**2).sum()
best_feature=min(gini_impurity,key=gini_impurity.get)
best_feature
```

```python
plt.figure(figsize=(15,10))
plot_tree(model,filled=True)
plt.title("Original Decision Tree")
plt.show()
```

#Lab 11
Prepare a model for prediction of prostate cancer using KNN Classifier


```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
df=pd.read_csv('prostate.csv')
print(df.head(10),"\n")
df.info()
```

```python
x=df.drop("Target",axis=1)
y=df["Target"]
print("Shape of x and y split dataset:",x.shape,y.shape)
sc=StandardScaler()
df1=pd.DataFrame(sc.fit_transform(x),columns=x.columns)
df1.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
knn_model=KNeighborsClassifier(n_neighbors=6)
knn_model.fit(x_train,y_train)
y_pred=knn_model.predict(x_test)
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("\nClassification Report:\n",classification_report(y_test,y_pred))
```

```python
error_rate=[]
for i in range(1,35):
  knn_model=KNeighborsClassifier(n_neighbors=i)
  knn_model.fit(x_train,y_train)
  new_y_pred=knn_model.predict(x_test)
  error_rate.append(np.mean(new_y_pred!=y_test)) #we are checking that the new pred is not the same as the previously derived pred and then taking its mean to append into list
plt.figure(figsize=(16,12))
plt.plot(error_rate,color="blue",linestyle="dashed",marker="o",markerfacecolor="red",markersize=10)
plt.title("Error Rate vs K Value")
plt.xlabel("K")
plt.ylabel("Error Rate")
plt.show()
```
